```python
import pandas as pd

# Create a sample DataFrame
data = {'Country': ['USA', 'USA', 'Canada', 'Canada'],
        'Indicator': ['GDP', 'Unemployment', 'GDP', 'Unemployment'],
        'Source': ['World Bank', 'BLS', 'World Bank', 'Statistics Canada'],
        'Year': [2020, 2020, 2020, 2020],
        'Value': [20000, 4.5, 17000, 7.0]}
df = pd.DataFrame(data)

# Create the MultiIndex
index = pd.MultiIndex.from_frame(df[['Country','Indicator','Source']])

# Set the MultiIndex as the DataFrame's index
df.index = index

# Drop the columns used to create the MultiIndex
df.drop(columns=['Country','Indicator','Source'],inplace=True)

# Print the resulting DataFrame
print(df)
```
